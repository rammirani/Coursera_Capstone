<h3>Model to predict the severity of an accident</h3>

Road traffic injuries are estimated to be the eighth leading cause of death globally and are predicted to become the seventh leading cause of death by 2030. Accidents in traffic lead to associated fatalities and economic losses every year and thus is an area of primary concern from loss prevention point of view.

The proactive approach includes a collision prevention approach, like, preventing a potential unsafe road conditions from occurring in the first place. Analysing a significant range of factors, including weather conditions, special events, roadworks, traffic jams among others, an accurate prediction of the severity of the accidents can be performed. 
These insights, could allow Government bodies to allocate their resources more effectively in advance, preventing when and where a severe accidents can occur as well as saving both, time and money. In addition, this knowledge of a severe accident situation can be warned to drivers so that they would drive more carefully or even change their route if it is possible or to inform hospitals - which could help anticipate the intervention in advance.

Modeling accident severity prediction and improving the model are critical to the effective performance of road traffic systems for improved safety.
This project is based on kaggle datasets that contain an extended descriptions of different aspect of the accidents, and the most relevant and useful data is considered for analysis.

Model involves predicting the Collision Severity. There are 3 classes under the Collision Severity {1,2,3}:
1 :  Fatal injury collision
2 :  Serious injury collision
3 :  Slight injury collision

The evaluation is the accuracy metric.
